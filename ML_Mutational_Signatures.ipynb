### Cancer Type Classification

Used the mutation data to implement a multi-class classification model with SVM and Neural Network and checked feature importance to identify which mutational channels are most predictive of specific cancer types.


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.feature_selection import SelectKBest, f_classif


data = somaticMutations.copy()


data['cancer_type'] = data.index.str.split('::').str[0]


X = data.drop('cancer_type', axis=1)
y = data['cancer_type']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

selector = SelectKBest(f_classif, k=20)  # You can adjust k
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)


selected_features = X.columns[selector.get_support()].tolist()
```

```python

svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train_selected, y_train)
svm_pred = svm_model.predict(X_test_selected)

print("SVM Classification Report:")
print(classification_report(y_test, svm_pred))


nn_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
nn_model.fit(X_train_selected, y_train)
nn_pred = nn_model.predict(X_test_selected)

print("\nNeural Network Classification Report:")
print(classification_report(y_test, nn_pred))
```

```python
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
r = permutation_importance(svm_model, X_test_selected, y_test, n_repeats=10, random_state=42)

feature_importance = pd.DataFrame({
    'feature': selected_features,
    'importance': r.importances_mean
})
feature_importance = feature_importance.sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(feature_importance['feature'], feature_importance['importance'])
plt.xticks(rotation=90)
plt.title('RBF SVM Feature Importance (Permutation)')
plt.tight_layout()
plt.show()
```

```python
nn_importance = np.mean(np.abs(nn_model.coefs_[0]), axis=1)
nn_feature_importance = pd.DataFrame({'feature': selected_features, 'importance': nn_importance})
nn_feature_importance = nn_feature_importance.sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(nn_feature_importance['feature'], nn_feature_importance['importance'])
plt.xticks(rotation=90)
plt.title('Neural Network Feature Importance')
plt.tight_layout()
plt.show()
```

### Graph-based Analysis:

Created a graph where nodes are samples and edges represent similarity above a threshold of 0.7 to detects communities using the Louvain method, calculates node centrality, and analyzed community sizes.
Identified which cancer types form the most central nodes
and analyzed intra-community connections to find most densely connected cliques that share similar signatures among them.


```python
import pandas as pd
import numpy as np
import networkx as nx
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
from community import community_louvain

similarity_matrix = cosine_similarity(mutation_data)

G = nx.Graph()

for sample in mutation_data.index:
    G.add_node(sample)


threshold = 0.7
for i in range(len(mutation_data)):
    for j in range(i+1, len(mutation_data)):
        if similarity_matrix[i, j] > threshold:
            G.add_edge(mutation_data.index[i], mutation_data.index[j], weight=similarity_matrix[i, j])

communities = community_louvain.best_partition(G)

centrality = nx.eigenvector_centrality(G)

pos = nx.spring_layout(G)
plt.figure(figsize=(12, 8))

nx.draw_networkx_nodes(G, pos, node_size=50, node_color=list(communities.values()))

nx.draw_networkx_edges(G, pos, alpha=0.1)

top_central_nodes = sorted(centrality, key=centrality.get, reverse=True)[:10]
labels = {node: node for node in top_central_nodes}
nx.draw_networkx_labels(G, pos, labels, font_size=8)

plt.title("Sample Similarity Graph with Communities")
plt.axis('off')
plt.tight_layout()
plt.show()

community_sizes = pd.Series(communities).value_counts().sort_index()
print("Community sizes:")
print(community_sizes)

print("\nTop 10 central nodes:")
for node in top_central_nodes:
    print(f"Sample: {node}, Centrality: {centrality[node]:.4f}, Community: {communities[node]}")

def analyze_connections(G, communities):
    intra_community = {i: 0 for i in set(communities.values())}
    inter_community = {i: 0 for i in set(communities.values())}

    for edge in G.edges():
        if communities[edge[0]] == communities[edge[1]]:
            intra_community[communities[edge[0]]] += 1
        else:
            inter_community[communities[edge[0]]] += 1
            inter_community[communities[edge[1]]] += 1

    return intra_community, inter_community

intra, inter = analyze_connections(G, communities)

print("\nIntra-community connections:")
print(intra)
print("\nInter-community connections:")
print(inter)

def analyze_community_mutations(mutation_data, communities):
    community_mutations = {}
    for community in set(communities.values()):
        community_samples = [sample for sample, comm in communities.items() if comm == community]
        community_mutations[community] = mutation_data.loc[community_samples].mean()
    return community_mutations

community_mutations = analyze_community_mutations(mutation_data, communities)

top_communities = community_sizes.nlargest(6).index
plt.figure(figsize=(15, 5*len(top_communities)))
for i, community in enumerate(top_communities):
    plt.subplot(len(top_communities), 1, i+1)
    community_mutations[community].plot(kind='bar')
    plt.title(f"Average Mutation Pattern for Community {community}")
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
